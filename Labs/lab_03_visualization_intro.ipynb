## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```

## Preview Data

Use the following data dictionary to understand the attributes available

|Attribute | Description |
|---------|--------------|
|	mpg	|Miles/(US) gallon|
|	cyl|	Number of cylinders|
|	disp|	Displacement (cu.in.)|
|hp|	Gross horsepower|
|drat|	Rear axle ratio|
|wt|	Weight (1000 lbs)|
|qsec|	1/4 mile time|
|vs|	Engine (0 = V-shaped, 1 = straight)|
|am|	Transmission (0 = automatic, 1 = manual)|
|gear|	Number of forward gears|


# Create basic charts

- Line chart
- Bar chart
- Histogram
- Box plot
- Scatter plot




```python
#Line chart
df.plot(x='carmodel', y='mpg', kind='line', title="MPG by Car")
```

```python
#Bar chart
df.plot(x='carmodel', y='mpg', kind='bar', title="MPG by Car")
```

```python
#Histogram
df['mpg'].plot(kind='hist', bins=10, title="MPG Distribution")
```

```python
#Boxplot
df['mpg'].plot(kind='box', title = 'Miles Per Gallon')
```

```python
#Scatter plot
df.plot.scatter(x='mpg',y='hp',c='wt')
```

```python
import matplotlib.pyplot as plt

counts = df['am'].value_counts()
labels = ['Automatic', 'Manual']

plt.figure()
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Transmission Type Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(df.index, df['hp'])
plt.xticks(rotation=90)
plt.xlabel('Car')
plt.ylabel('Horsepower')
plt.title('Horsepower by Car')
plt.show()
```

```python
plt.figure()
plt.scatter(df['wt'], df['mpg'])
plt.xlabel('Weight')
plt.ylabel('Miles Per Gallon')
plt.title('Weight vs MPG')
plt.show()
```

```python
avg_mpg = df.groupby('cyl')['mpg'].mean()

plt.figure()
avg_mpg.plot(kind='bar')
plt.xlabel('Cylinders')
plt.ylabel('Average MPG')
plt.title('Average MPG by Cylinder')
plt.show()
```

```python
plt.figure()
plt.hist(df['wt'], bins=10)
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Weight Distribution')
plt.show()
```
